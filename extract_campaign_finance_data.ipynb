This script downloads all the donation information from the Wisconsin governmental website onto individual html files. 
I filtered the results to contain the information from the donations made between November 8, 2015 and November 8, 2016. Donation information from each page of the filtered results will be written into individual HTML files. Each HTML file will contain 100 entries. 

```python
#import necessary packages 
import os
import requests
import re
```

```python
#make a category called campaign finance 
category = "campaign_finance"
```

```python
#make a new folder in the final_project folder called 'campaign_finance'. This folder will contain all scrapped html files from 
#Wisconsin governmental website

if os.path.exists('Users/jding/CSB/python/final_project/') == False:
    os.system('mkdir -p /Users/jding/CSB/python/final_project/' + category)
```

```python
#Url for the first page of the filtered search. search is filtered by date- Nov 8, 2015 - Nov 9, 2016
url = 'http://www.wisdc.org/index.php?from=2015-11-08&to=2016-11-08&filter=+Search+&module=wisdc.websiteforms&cmd=searchadvanced&qty=100' + '&p=1'
first_page = requests.get(url, allow_redirects = True)
```

```python
first_page.content
```

```python
#Write html files for entire date range of filtered results. Each html file contains 100 entries.  
for i in range(1, 464): 
    print('downloading page ', i, 'in category ', category)
    url = 'http://www.wisdc.org/index.php?from=2015-11-08&to=2016-11-08&filter=+Search+&module=wisdc.websiteforms&cmd=searchadvanced&qty=100' +  '&p=' + str(i)
    print(url)
    page = requests.get(url, allow_redirects = True)
    with open('/Users/jding/CSB/python/final_project/' + category + "/" + str(i) + ".html", 'wb') as f: 
        f.write(page.content)
```
